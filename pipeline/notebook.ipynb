```python
import pandas as pd
```

```python
prefix = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow'
url = f'{prefix}/yellow_tripdata_2021-01.csv.gz'
df = pd.read_csv(url)
# .gz = compressed file
```

```python
# format the data types
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates
)
```

```python
from sqlalchemy import create_engine

#create data connection
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
df.head(0)
```

```python
# create table in postgresql
# df.head(0) takes only the column structure
df.head(0).to_sql(name="yellow_taxi_data", con=engine, if_exists='replace')
```

```python
# scheme installed inside the database
# print scheme
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
# load data 100000 rows per chunk
df_iter = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000
)
```

```python
df_iter
```

```python
!uv add tqdm
```

```python
from tqdm.auto import tqdm
#tqdm for seeing the process of data chrunk
# data chrunk to prevent memory problem
# load data into the table
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(name="yellow_taxi_data", con=engine, if_exists='append')
```
